# Соединяем таблицы по ['Материал', 'МРЦ'], выкидываем nan'ы по МРЦ

```python
df3 = df1.merge(df2, how = 'inner', left_on = ['Материал', 'МРЦ'], right_on = ['Материал', 'МРЦ'])
```

```python
df3.columns
```

```python
df3.dropna(subset=['МРЦ'], inplace = True)
```
